```python
import pandas as pd
```

```python
df = pd.read_csv("../processed/bps_with_doses_old.csv")
```

```python
import seaborn as sns
from matplotlib.pyplot import figure
figure(figsize=(12,8), dpi=80)
df_filtered = df[~df["dose"].isna()]
df_filtered = df[(df["cur_bp"]<200) & (df["cur_bp"]>0) & (df["dose"]<0.4)]
sns.lineplot(data=df_filtered, x="cur_bp", y="dose")
```

```python
figure(figsize=(12,8), dpi=80)
df_filtered_50_100 = df[~df["dose"].isna()]
df_filtered_50_100 = df[(df["cur_bp"]<100) & (df["cur_bp"]>50) & (df["dose"]<0.4)]
sns.lineplot(data=df_filtered_50_100, x="cur_bp", y="dose")
```

```python
figure(figsize=(12,8), dpi=80)
i = 0
for subject_id in df_filtered_50_100["subject_id"].unique():
    subset_df = df_filtered_50_100[df_filtered_50_100["subject_id"] == subject_id]
    sns.lineplot(data=subset_df, x="cur_bp", y="dose", color="blue", estimator=None, lw=0.1)
    if i > 1000:
        break
    i += 1
```
